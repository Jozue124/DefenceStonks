```python
import pandas
import os
import matplotlib as plt
import re
```

# Analysis of defence companies stocks over time

## Reading in files from csv into 1 "nice" dataframe

```python
#change directory
os.chdir("archive")
```

### reading in datasets 

```python
#datasets array we gonna combine after
datasets = []
for filename in os.listdir(os.getcwd()):
    if filename.endswith(".csv"):
        #reading in file
        temp = pandas.read_csv(filename)
        #use regex to get comapny name
        reg = re.search(r"^(.+)(Historical Data.csv)",filename)
        companyName = reg.group(1)
        #make array with company name
        company = [companyName]*temp['Date'].size
        #make a new collumn in df for company name
        temp["company"] = company
        #add to array of dataframes
        datasets.append(temp)
    else:
        print("ERROR")
```

```python
#so dataframe has all the info but datasets is still useful
dataframe = pandas.concat(datasets)
```

## visualization pt1

```python
#plotting all closing values
#params
plt.rcParams["figure.figsize"] = (50,10)
plt.rcParams["figure.autolayout"] = True
#first plt
ax = datasets[0].plot(x='Date', y='Close')
#all toher plots 
for ds in range(1,len(datasets)-1):
    datasets[ds].plot(ax=ax, x='Date', y='Close')
```
